```python
# 맨 앞과 뒤의 빈 리스트 삭제
print(data[0], data[-1])
del data[0], data[-1]

# 순위, SV, HLD 삭제
print(data[0][0], *data[0][7:9])

for row in data:
    del row[7:9]
    del row[0]
```

```python
# 처리결과 확인하기
print(*data[0:3], sep='\n')
```

```python
# ERA, G, W 등의 타입을 str에서 데이터에 맞게 바꾸기
for row in data[1:]:
    row[2], row[6], row[15] = float(row[2]), float(row[6]) , float(row[15])

    # 리스트 컴프리헨션으로 리스트 대치
    row[3:6] = [int(x) for x in row[3:6]]
    row[8:15] = [int(x) for x in row[8:15]]

    # IP는 따로 처리 
    ip = row[7].split() # row[7]을 공백을 기준으로 분리
    if len(ip) == 2:    # 분리된 문자열 리스트의 길이가 2 == 이닝이 소수점
        # 이닝별 소수 처리
        if ip[1] == '1/3':
            ip[1] = 0.1
        elif ip[1] == '2/3':
            ip[1] = 0.2
        row[7] = float(ip[0])+ip[1] # 처리된 이닝을 더해줌
    else:
        row[7] = float(row[7]) # 소수 부분이 없으면 타입 캐스팅만 실행
```

```python
# 처리결과 확인하기
print(*data[:6], sep='\n')
```

```python
# 이닝 순으로 내림차순 정렬
sort_by_IP = sorted(data[1:], key=lambda x: x[7], reverse=True)
sort_by_IP.insert(0, data[0])
print(*sort_by_IP[0:6], sep='\n')

print()

# 볼넷 순으로 오름차순 정렬
sort_by_BB = sorted(data[1:], key=lambda x: x[10])
sort_by_BB.insert(0, data[0])
print(*sort_by_BB[0:6], sep='\n')
```

```python
# 함수 사용해보기
# 선수 이름으로 검색하는 함수
def search_by_name(name, data):
    for player in data[1:]:
        if player[0] == name:
            return player
    return None

while True:
    name = input().strip()
    result = search_by_name(name, data)
    if result is None:
        print("선수 이름을 확인해주세요.")
    else:
        print(data[0])
        print(result)
        break
```
