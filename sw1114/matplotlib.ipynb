## Matplotlib

Matplotlib으로 그릴 수 있는 그래프들

- https://matplotlib.org/stable/gallery/index.html


```python
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
%matplotlib inline
```

`plt.plot()`

- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

> 함수원형( 대괄호[ ] 쳐진 요소들은 생략 가능 )

![plot](img/plot.png)

- [x], y: x값, y값. 
    - x값 생략시 y값에 맞게 생성

- [fmt]: format string. 포맷 문자열로 그래프의 스타일 설정

- *: 이 뒤에 오는 인수들은 놓인 순서로 판단하지 않고 data=None처럼 키워드를 꼭 써야 함을 표시

- data: 데이터 들어있는 딕셔너리, 데이터프레임을 객체로 지정해 그래프 그리기

- **kwargs: keyword argument, 추가적인 키워드 인수. 그래프 라인 색, 두께, 스타일, 라벨 등을 지정


`plt.show()`
- 그래프를 화면에 그리기

```python
# 간단한 예시용 랜덤 리스트

rd = []
for i in range(10):
    r = random.randrange(5)
    rd.append(r)
print(rd)
```

```python
# 랜덤 리스트로 시리즈 만들기
sd = pd.Series(rd)
sd
```

```python
#plot()에 매개변수 한개를 써주면 y값으로 처리
plt.plot(sd) # plt.plot([랜덤 리스트]) 와 같다.
plt.show()
```

```python
# plot(x, y) x, y 모두 지정해주기
plt.plot([1,2,3,4,5],[1,4,3,5,6])
plt.show()
```

```python
# x축 라벨 지정하기 
month = ['aug', 'oct', 'nov',' dec']
nums = [2, 5, 6, 4]
plt.plot(month, nums)
plt.show()
```

```python
# x, y 데이터의 개수 맞지 않으면 오류 발생
plt.plot([1,2,3],[1,4,3,5,6])
```

```python
# 라벨링된 데이터 입력받아 그래프 그리기
# 함수원형의 data=None 부분
labelData = {'x':[1,2,3,4,5], 'y':[1,2,5,4,3]}
obj = pd.DataFrame(labelData)
obj
```

```python
# 라벨링된 데이터 입력받아 그래프 그리기
# 이 경우엔 x,y라벨을 지정해야 함
plt.plot('x', 'y', data=obj)
plt.show()
```

### 포맷 스트링 설정
![fmt](img/fmt.png)

```python
# 그래프 스타일1, [fmt] 설정하기
# fmt = '[marker][line][color]' or '[color][marker][line]
plt.plot(sd, 'o:r') # 마커 o, 라인 :(dot), 컬러 red
plt.show()
```

```python
# 그래프 스타일2, **kwargs 사용해서 색 설정
plt.plot(sd,'*--', color='#0A4A9B') # 마커 *, 라인 --, 컬러 #0a4a9b
plt.show()
```

```python
# x, y축 범위 지정하기
plt.plot(sd, 'd--r')
# plt.xlim(0, 9) # x축 범위 지정하기
# plt.ylim(0, 5) # y축 범위 지정하기
plt.axis([0, 9, 0, 5]) # x, y축 한번에 지정하기 [xmin, xmax, ymin, ymax]
plt.show() 
```

```python
# fill_between(), 그래프 영역 채우기
x = range(5)
y = [3, 2, 2, 5, 4]

plt.plot(y)
plt.fill_between(x[1:3],y[1:3])
plt.show()
```

### title, legend, label, grid
**title**은 그래프 제목 설정

**legend**는 범례 지정

- 범례: 각 그래프의 의미를 설명하는 라벨

**label**로 각 축의 이름 설정 가능

**grid**로 격자선 출력


```python
plt.rc('font', family='Pretendard variable') # 한글폰트 지정

plt.title("타이틀 교체") # 타이틀 지정
plt.xlabel("x 라벨") # x, y 라벨 지정
plt.ylabel("y 라벨")

plt.plot(sd,'-g', label="라벨") # 그래프 라벨 지정
# plt.legend() # 그래프 범례 그리기
plt.legend(loc='upper left') # 그래프 범례 위치 지정 location (좌상단으로)
plt.grid() # 격자 출력
plt.show()
```

```python
# 그래프 두개 그리기
x = ['A', 'B', 'C', 'D', 'E']
data1 = [10, 30, 20 ,40, 50]
data2 = [30, 20, 40, 60, 20]

plt.plot(x, data1, label='data1')
plt.plot(x, data2, label='data2')
plt.legend(loc='upper left')
plt.show()
```

## 꺾은선 그래프 그리기 종합

sin, cos 그래프 그리기


```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(10, 6)) # 그래프 크기 가로 10인치, 세로 6인치로 설정

plt.plot(x, y, x, z) # plt.plot() 두 개 한번에 그리기
plt.title("사인, 코사인 그래프") # 타이틀 설정
plt.xlabel('x: 0 ~ 4pi') # x, y 라벨 설정
plt.ylabel('y: sin(x) & cos(x)')

# x, y 틱 설정하기 (눈금 표시)
# 정해진 x값을 labels로 표시
# xticks([눈금 위치 값], labels=[표시할 값])
plt.xticks([np.pi, np.pi*2, np.pi*3, np.pi*4],labels=['π','2π','3π','4π'])
plt.yticks([-1, -0.5, 0, 0.5, 1])


plt.grid(True, which='major', color='gray', linestyle='--') # 격자 옵션 설정해서 그리기
# True = 격자 보이게 설정
# which = 표시할 격자선 종류 지정, both, major, minor
# color = 컬러 지정
# linestyle = 라인 스타일 지정

plt.legend(['sin(x)', 'cos(x)']) #plot에서 label 설정 없이 legend로 바로 범례 추가
plt.show()
```

## 막대 그래프 그리기

`plt.bar(x, y)`

- 막대 그래프는 x, y 둘 다 넣어야 함

```python
x = range(1,6)
y = [1, 3, 5, 4, 5]
plt.title("막대 그래프")
plt.bar(x, y)
plt.show()
```

```python
# 막대 눕히기, plt.barh(horizontal)
plt.barh(x, y, color='green', alpha=0.3)
plt.show()
```

```python
# 히스토그램 그리기
# 데이터 빈도 표시

h = []
for i in range(10):
    h.append(random.randint(1,10))

plt.hist(h)
plt.show()
```

```python
h = np.random.randn(10000) # 평균 0, 표준편차 1인정규분포 데이터 10000개 생성
h2 = np.random.randn(10000) * 2 # 평균 0, 표준편차 2인 정규분포 데이터 10000개 생성

plt.hist(h,color='#9dae11', bins=100, alpha=0.8, density=True, label='data1')
plt.title('histtype =  bar')
# color = 색 지정
# bins = 구간의 개수 지정 (50칸으로 나눔)
# alpha = 투명도 (0 ~ 1)
# histtype = 히스토그램 타입 (bar, barstacked, step, stepfilled) default='bar'
# density = True시 y축이 데이터 개수가 아니라 확률밀도로 표시됨

plt.hist(h2,color='#0A4A9B', bins=100, alpha=0.9, histtype='step', density=True, label='data2')
plt.title('histogram')
plt.legend()

plt.show()
```

## 산점도 그리기

`plt.scatter(x, y)`

- 점으로 분포도를 알 수 있는 산점도 그래프



```python
# 랜덤 값 산점도
x = [random.randint(5,11) for _ in range(10)]
y = [random.randint(1,10) for _ in range(10)]

plt.scatter(x, y, color='#0a4a9b', alpha=0.8)
# 기존에 사용했던 color, alpha옵션들 사용 가능
plt.show()
```

```python
# 상관관계 있는 데이터 생성 (예시)
np.random.seed(0)
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # 공분산 행렬로 양의 상관관계 설정
x, y = np.random.multivariate_normal(mean, cov, 100).T


# 추세선 그리기
# 1차 다항식 (y = ax + b)에 대한 계수 추정
slope, intercept = np.polyfit(x, y, 1)

plt.plot(x, slope * x + intercept, color='#FF5733',alpha=0.95, label="추세선 (양의 상관관계)", linewidth=2 )

# 산점도 그리기
plt.scatter(x, y, label="데이터 점", color='#0a4a9b')
plt.xlabel("X")
plt.ylabel("Y")

# 제목과 범례 추가
plt.title("양의 상관관계 산점도 그래프")
plt.legend()
plt.grid(True, color='gray', alpha=0.5)
plt.show()
```
