## 데이터 추가

```python
df['수학 성적']=[97,85,45,94,77,60,88]
df['학번']=[1111,1112,2222,2223,3311,3322,3333]
df=df.set_index('학번') #인덱스 변경
df
```

```python
# column 순서 변경
df1=df[['학년','성별','수학 성적','영어 성적']]
df1
```

```python
df['영어 등급'] ='F'  #새로운 Column 결과를 만들고 전체 데이터를 F로 초기화
df
```

```python
df.loc[df['영어 성적'] >= 90, '영어 등급'] = 'A'
df.loc[(df['영어 성적'] >= 80) & (df['영어 성적'] < 90), '영어 등급'] = 'B'
df.loc[(df['영어 성적'] >= 70) & (df['영어 성적'] < 80), '영어 등급'] = 'C'
df
```

```python
import numpy as np
df['영어 등급'] = np.where(df['영어 성적'] >= 90, 'A', 
                   np.where(df['영어 성적'] >= 80, 'B', 
                   np.where(df['영어 성적'] >= 70, 'C', 'F')))
df #np.where: R의 ifelse처럼 ~가 아니면 !, !가 아니면 *식으로 작동하는 함수
```

```python
#upper() 대문자  lower() 소문자
df['영어 등급']=df['영어 등급'].str.lower()
df
```

```python
df['영어 등급']=df['영어 등급'].str.upper()
df
```

## 데이터 그룹화

```python
df.groupby('Sex').size() #성별 개수 카운트
```

```python
df.groupby('Grade')['Math_score'].mean() #학년으로 그룹화 -> 수학 성적 평균
```

```python
df.groupby('Grade')[['Eng_score','Math_score']].mean() #학년별 영어, 수학 성적 평균
```

```python
df.groupby(['Grade','Sex'])[['Eng_score','Math_score']].mean()
#학년별, 성별에 따른 영어, 수학 성적 평균 데이터

df.groupby(['Grade','Sex'])[['Eng_score','Math_score']].mean().sort_values('Eng_score')
#영어 성적 오름차순 정렬

df.groupby(['Grade','Sex'])[['Eng_score','Math_score']].mean().sort_values('Eng_score',ascending=False)
#영어 성적 내림차순 정렬
```

```python
df.groupby('Grade')[['Math_score','Eng_score']].sum() #학년별 수학, 영어 성적 합계
```
